```python
#TASK=2
class Shape:
  def __init__(self, s, b, h) :
    self.shape=s
    self.base=b
    self.height=h
  def area(self):
    if self.shape =="Triangle" :
      area = 0.5 * self.base *self.height
      print(f'Area: {area}')
    elif self.shape =="Square" :
      area=self.base*self.height
      print(f'Area: {area}')
    elif self.shape == "Rhombus" :
      area = 0.5*self.base*self.height
      print(f'Area: {area}')
    elif self.shape =="Rectangle":
      area=self.base*self.height
      print(f'Area: {area}')
    else:
      print(f'Area: Shape unknown ')

triangle=Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#TASK=22
class Hotel():
    def __init__(self, hn):
        self.hname = hn
        self.stuffcount = 0
        self.guestcount = 0
        self.stuff = {}
        self.guest = {}

    def addStuff(self, stn, sage, ph= "000"):
        self.stuffcount += 1
        self.stuff[self.stuffcount] = [stn, sage, ph]
        print(f"Staff with ID {self.stuffcount} is created.")

    def getStuffById(self, n):
        return f"Staff Id: {n}\n Name: {self.stuff[n][0]}\nAge: {self.stuff[n][1]}\nPhone: {self.stuff[n][2]}"

    def addGuest(self, gn, gage, ph):
        self.guestcount += 1
        self.guest[self.guestcount] = [gn, gage, ph]
        print(f"Guest with ID {self.guestcount} is added.")

    def getGuestById(self, n):
        return f"Guest Id: {n}\nName: {self.guest[n][0]}\nAge: {self.guest[n][1]}\nPhone: {self.guest[n][2]}"

    def allGuest(self):
        print("All Guest: ")
        print(f"Number of all guest: {self.guestcount}")
        for k, v in self.guest.items():
            print(f"Guest ID: {k} Name: {v[0]} Age: {v[1]} Phone no.: {v[2]}")

    def allStaffs(self):
        print("All Staff: ")
        print(f"Number of all Staff: {self.stuffcount}")
        for k, v in self.stuff.items():
            print(f"Staff ID: {k} Name: {v[0]} Age: {v[1]} Phone no.: {v[2]}")


# Do not change the following lines of code.
h = Hotel("Lakeshore")
h.addStuff("Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest("Carol", 35, "123")
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, "431")
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```

```python
# Task 23
class Author():
    def __init__(self, name=None):
        self.name = name
        self.book = {}
        self.bname = ""
        self.gname = ""
        self.booknumber = 0

    def setName(self, newname):
        self.name = newname

    def addBook(self, b, g):
        self.bname = b
        self.gname = g
        if self.name == None:
            print("A book can not add without an author name.")
        else:
            self.booknumber += 1
            if self.gname not in self.book:
                self.book[self.gname] = [self.bname]
            else:
                self.book[self.gname].append(self.bname)

    def printDetail(self):
        print(f"Number of Book(s): {self.booknumber}")
        print(f"Author Name: {self.name}")
        for k, v in self.book.items():
            print(k, ":", str(v)[1:-1])


# Do not change the following lines of code.
a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print( "== == == == == == == == == == == == == == == ===")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```

```python
# Task 24
class Doctor():
    def __init__(self, id, prof, n, spc):
        self.id = id
        self.profession = prof
        self.name = n
        self.speciality = spc

class Patient():
    def __init__(self, pid, r, pn, ag, ph):
        self.pid = pid
        self.pname = pn
        self.age = ag
        self.phone = ph

class Hospital():
    def __init__(self, hname):
        self.Hname = hname
        self.docinfo = {}
        self.patinfo = {}

    def addDoctor(self, docobj):
        key = docobj.id
        value = [docobj.name, docobj.speciality]
        self.docinfo[key] = value

    def addPatient(self, patobj):
        key = patobj.pid
        value = [patobj.pname, patobj.age, patobj.phone]
        self.patinfo[key] = value

    def getDoctorByID(self, idstr):
        output = "Doctor ID: "+ idstr
        dname, dspc = self.docinfo[idstr]
        output += f"\nNmae: {dname} \nSpeciality: {dspc}"
        return output

    def getPatientByID(self, pidstr):
        output = "Patient ID: "+ pidstr
        pname, age, phone = self.patinfo[pidstr]
        output += f"\nPatient Nmae: {pname} \nAge: {age} \nPhone no: {phone}"
        return output

    def allDoctors(self):
        print("All Doctor: ")
        print(f"Number of all Doctor: {len(self.docinfo)}")
        print(self.docinfo)

    def allPatients(self):
        print("All Patients: ")
        print(f"Number of all patients: {len(self.patinfo)}")
        print(self.patinfo)


# Do not change the following lines of code.
h = Hospital("Evercare")
d1 = Doctor("1d", "Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p", "Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient("2p", "Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
#TASK=25
class Vaccine():
    def __init__(self, vn, con, d):
        self.vname = vn
        self.country = con
        self.next = d

class Person():
    def __init__(self, n, ag, st="General Citizen"):
        self.pname = n
        self.age = ag
        self.status = st
        self.vacc1 = None
        self.vacc2 = None
        self.nextDate = 0
        self.pvname = ""

    def pushVaccine(self, vacobj, dos="1st dose"):
        if self.status != "Student":
            if self.age < 25:
                print(f"Sorry {self.pname}, minimum age for taking vaccine is 25 years now.")

        if dos == "1st dose":
            self.vacc1 = "Given"
            print(f"{dos} is done for {self.pname}")
            self.pvname = vacobj.vname
            self.nextDate = vacobj.next
        else:
            self.vacc2 = "Given"
            print(f"{dos} is done for {self.pname}")

        if self.pvname != vacobj.vname:
            print(f"Sorry {self.pname}, you can not take 2 different vaccine")

    def showDetail(self):
        print(f"Name: {self.pname} Age: {self.age} Type: {self.status}")
        print(f"Vaccine Name: {self.pvname}")
        print(f"1st Dose: {self.vacc1}")
        if self.vacc2 == None:
            print(f"2nd Dose: Please come after {self.nextDate} days.")
        else:
            print(f"2nd Dose: {self.vacc2}")


# Write your code here
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
